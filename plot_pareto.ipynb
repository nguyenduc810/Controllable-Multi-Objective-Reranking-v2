```python
def min_max_scaling_along_axis1(data):
    min_vals = np.min(data, axis=1, keepdims=True)
    max_vals = np.max(data, axis=1, keepdims=True)
    normalized_data = (data - min_vals) / (max_vals - min_vals)
    return normalized_data
```
